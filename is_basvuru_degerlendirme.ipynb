# Kütüphanelerin Import Edilmesi

```python
import numpy as np
import pandas as pd
from sklearn import tree
```

```python
df = pd.read_csv("DecisionTreesClassificationDataSet.csv")
df
```

# Mapping İşlemlerinin Yapılması

```python
duzetme_mapping = {'Y': 1, 'N': 0}

df['IseAlindi'] = df['IseAlindi'].map(duzetme_mapping)
df['SuanCalisiyor?'] = df['SuanCalisiyor?'].map(duzetme_mapping)
df['Top10 Universite?'] = df['Top10 Universite?'].map(duzetme_mapping)
df['StajBizdeYaptimi?'] = df['StajBizdeYaptimi?'].map(duzetme_mapping)
duzetme_mapping_egitim = {'BS': 0, 'MS': 1, 'PhD': 2}
df['Egitim Seviyesi'] = df['Egitim Seviyesi'].map(duzetme_mapping_egitim)
df.head()
```

# X ve Y değerlerinin Ayrıştırılması

```python
y = df[['IseAlindi']]
X = df.drop(['IseAlindi'], axis=1)
```

```python
X.head()
```

# Decision Tree Algoritmasının Uygulanması

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X.values,y) #Sadece X olarak uyguladığımda değişken adlarını kaldırmamız gereken bir uyarı veriyor.
```

# Uygulamanın Fonksiyona Çevrilmesi

```python
def human_resources():
    basvuran=[]
    basvuran.append(input("Başvuranın deneyim yılı(sayı): "))
    basvuran.append(input("Başvuran çalışıyor mu?(Hayır:0, Evet:1) "))
    basvuran.append(input("Başvuranın eskiden çalıştığı firma sayısı: "))
    basvuran.append(input("Başvuranın eğitim seviyesi(BS:0, MS:1, PhD:2): "))
    basvuran.append(input("Başvuran top 10 üniversiteden birinde okumuş mu?(Hayır:0, Evet:1) "))
    basvuran.append(input("Başvuran şirketimizde staj yaptı mı?(Hayır:0, Evet:1) "))
    if (clf.predict([basvuran])) == 0:
        print("Başvuran şirket kriterlerine uygun değildir.")
    else:
        print("Başvuran şirket kriterlerine uygundur.")

    
```
